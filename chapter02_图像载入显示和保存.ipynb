# 基本元素-图片载入显示和保存

## 目标
* 加载、显示和保存图片
* opencv函数：cv.imread(),cv2.imshow(),cv2.imwrite()

## 正文

电脑彩色图是以RGB颜色模式显示得，而opencv中彩色图是以B-G-R通道顺序存储的。图像坐标起始点为左上角，行对于y,列对应x:

![](http://jupter-oss.oss-cn-hangzhou.aliyuncs.com/public/files/image/1095279207225/1546838566157_BHzKl3QtQe.jpg)

### 加载图片
函数：cv2.imread()

````python
import cv2
#读取灰度图像
img=cv2.imread('lena.jpg',0)
````

* 参数1：文件名
* 参数2：读取模式
```
cv.IMREAD_COLOR:彩色图，默认值（1）
cv.IMREAD_GRAYSCALE:灰度图(0)
cv.IMREAD_UNCHANGED:包含透明通道得彩色图（-1）
```

注： 路径中不能有中文噢，并且没有加载成功的话是不会报错的

### 显示图片
函数：cv2.imshow()

````python
cv2.imshow('lena',img)
cv2.waitKey(0)
````

* 参数1：显示窗口得名字，不同窗口之间用窗口名区分，所以窗口名相同就表示是同一个窗口
* 参数2：要显示得图片


cv2.waitKey()是让程序暂停的意思，参数是等待时间（毫秒ms）。时间一到，会继续执行接下来的程序，传入0的话表示一直等待。等待期间也可以获取用户的按键输入：k = cv2.waitKey(0)

![](http://jupter-oss.oss-cn-hangzhou.aliyuncs.com/public/files/image/1095279207225/1546838882589_RkRHZnWrBh.jpg)

还可以先用cv2.namedWindow()创建一个窗口，之后显示图片

````python
cv2.namedWindow('lena2',cv2.WINDOW_NORMAL)
cv2.imshow('lena2',img)
cv2.waitKey(0)
````

* 参数1：窗口名
* 参数2：默认为cv2.WINDOW_AUTOSIZE 表示窗口大小自动适应图片大小，也可以设置为WINDOW_NORMAL:表示窗口大小可以调整

### 保存图片
函数cv2.imwrite()

````python
cv2.imwrite('lena_gray.jpg',img)
````

* 参数1：要保存得文件名，包含图片格式后缀
* 参数2：要保存得图片

## 练习

* (1) 打开lena.jpg并显示，如果按下’s’，就保存图片为’lena_save.bmp’，否则就结束程序。

````python
import cv2
img=cv2.imread('lena.jpg')
cv2.imshow('lena',img)
k=cv2.waitKey(0)
if k==ord('s'):
    cv2.imwrite('lena_save.bmp',img)
````
